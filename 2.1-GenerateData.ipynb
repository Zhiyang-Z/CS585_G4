```python
#PART 1

#https://numpy.org/doc/2.1/reference/random/generated/numpy.random.uniform.html

import numpy as np
import pandas as pd
np.random.seed(42) #for reproducibility

#a dataset of at least 6,000 points
pointsNum = 6000

#w values range from 0 to 10,000
w = np.random.uniform(0, 10000, pointsNum)
#x values range from 20,000 to 1,000,000
x = np.random.uniform(20000, 1_000_000, pointsNum)
#y and z can be whatever you decide-selected to generate random numbers between 0 and 1 for y and z
y = np.random.uniform(0, 1, pointsNum)
z = np.random.uniform(0, 1, pointsNum)

data = pd.DataFrame({
    'w': w,
    'x': x,
    'y': y,
    'z': z,
})

data.to_csv("kmeansdataset.csv", index=False)
print("Dataset generated.")
```

```python
#PART 2
import numpy as np
import pandas as pd

np.random.seed(42) #for reproducibility

#k initial seed points
K=5

#K random seed points
seed_w = np.random.uniform(0, 10000, K)
seed_x = np.random.uniform(20000, 1_000_000, K)
seed_y = np.random.uniform(0, 1, K)
seed_z = np.random.uniform(0, 1, K)


seed_data = pd.DataFrame({
    'w': seed_w,
    'x': seed_x,
    'y': seed_y,
    'z': seed_z,
})


seed_data.to_csv("kmeans_seeds.csv", index=False)
print(f"{K} seed points generated.")
```
